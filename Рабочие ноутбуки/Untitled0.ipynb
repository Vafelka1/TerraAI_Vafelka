```python
import os
```

```python
for i in os.listdir('1_Метки'):
  with open(f'1_Метки/{i}', 'r+') as f:
    text = f.readline()
    nt = str(text.split(' ')[1])+' '+str(text.split(' ')[3])+' '+str(text.split(' ')[2])+' '+str(text.split(' ')[4])
    new_text = f'0 {nt}'
    f.truncate(0)

  with open(f'1_Метки/{i}', 'w') as f:
    f.write(new_text)
```
